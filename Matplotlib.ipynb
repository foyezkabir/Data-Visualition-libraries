```python
#importing matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# parabola

x=np.linspace(-10,10,20)
y= x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
x=np.linspace(-5,5,50)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'r*')
plt.plot(x,z,'g-')
```

```python
#bar plot
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
languages= ('bangla','hindi','german','english','french')
people= (50,150,40,60,90)
ax.bar(languages, people)
plt.xlabel('languages')
plt.ylabel('people')
plt.title('nibba bois')
plt.show()
```

```python
#scater plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
```

```python
# 3D scatter plot
fig1=plt.figure()
ax=plt.axes(projection='3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

```python
#scatter plot
x=np.linspace(-5,6,20)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
```

```python
#pie chart
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
city=('Rajshahi','Dhaka','Nougha','Kurigram','Bandorban')
fruits=(100,90,40,70,50)
ax.pie(fruits, labels=city, autopct='%1.1f%%')
plt.show()
```

```python
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
city=('Rajshahi','Dhaka','Nougha','Kurigram','Bandorban')
fruits=(80,100,40,70,50)
ax.bar(city, fruits)
plt.xlabel('Fruits')
plt.ylabel('City')
plt.title('Fruits in the City')
plt.show()
```
