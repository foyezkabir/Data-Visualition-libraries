```python
#loading house price dataset
from sklearn.datasets import load_boston
house_boston= load_boston()

house = pd.DataFrame(house_boston.data, columns= house_boston.feature_names)
house['PRICE']=house_boston.target
```

```python
print(house_boston)
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Correlation-- it says which colum is important for our prediction,and which r not 

   --Positive
   *italicised text*
   --Negative 

```python
#finding correlation
correlation = house.corr()
```

Heat Map

```python
#contrusting a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square= True, fmt= '.1f', annot= True, annot_kws={'size':8},cmap='Reds')
```
